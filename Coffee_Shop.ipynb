# Coffee Shop BI

This project was something I did strictly using Tableau to play around with data. I got the idea from a handful of different blog posts I had seen come through my inbox from Medium.

The first thing I looked at was how many products were sold at what time, from opening to closing of all of the locations. Then I split that by location of the coffee shop (three separate locations).

The product category that sold the most was unsurprisingly coffee. This is a coffee shop, afterall. Tea & baked goods were the next two highest selling products.

Focusing on coffee, tea, & bakery goods would be priority since those are the products that get sold the most. Certain non-edible products, like t-shirts & mugs, might be nice to supply more of around the holidays (winter & summer vacations).

The next thing I took a peek at was the total quantity of products sold and at what time per location.

Each location had a similar trend, which was also shown on the first graph: most products were sold earlier in the day.

From a staff management prespective, it's recommended to have more staff members working earlier in the day. More orders means more staff will be needed, whether that be at cash registers taking orders, handing out pastries, or working the espresso machine(s) & making drinks.
